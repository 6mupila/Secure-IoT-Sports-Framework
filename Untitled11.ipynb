```python
import pandas as pd

# Load your real dataset
real_data = pd.read_csv("DSL-StrongPasswordData.csv")
display(real_data.head())
```

```python
from faker import Faker
fake = Faker()

imposter_data = []

for _, row in real_data.iterrows():
    imposter_row = row.copy()
    for col in real_data.columns[3:]:  # Skip metadata columns
        noise = fake.random_number(digits=3, fix_len=True) / 100.0
        noise *= fake.random_element(elements=[-1, 1])
        imposter_row[col] += noise
    imposter_data.append(imposter_row)

imposter_df = pd.DataFrame(imposter_data)
```

```python
combined_data = pd.concat([real_data, imposter_df], ignore_index=True)
display(combined_data.head())
```

```python
combined_data.to_csv("combined_keystroke_data.csv", index=False)
print("Combined dataset saved as 'combined_keystroke_data.csv'")
```
