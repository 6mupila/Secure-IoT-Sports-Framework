```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/mnt/data/Data_All_221114 (2).pdf"

# Since PDFs cannot be directly read into DataFrames, summarizing key findings manually from extracted text

# Cybersecurity awareness levels
awareness_levels = {
    "Extremely aware": 68,
    "Very aware": 81,
    "Somewhat aware": 42,
    "Not so aware": 15,
    "Not at all aware": 2
}

# Convert to DataFrame for visualization
awareness_df = pd.DataFrame(list(awareness_levels.items()), columns=["Awareness Level", "Count"])

# Create a bar plot for awareness levels
plt.figure(figsize=(8, 5))
sns.barplot(x="Count", y="Awareness Level", data=awareness_df, palette="Blues_r")
plt.xlabel("Number of Respondents")
plt.ylabel("Cybersecurity Awareness Level")
plt.title("Cybersecurity Awareness Levels Among Respondents")
plt.show()

# Internet usage activities
internet_activities = {
    "Social Media": 162,
    "Entertainment": 93,
    "Research": 121,
    "Online Learning": 112,
    "Job Opportunities": 83,
    "Other": 7
}

# Convert to DataFrame for visualization
activities_df = pd.DataFrame(list(internet_activities.items()), columns=["Activity", "Count"])

# Create a pie chart for internet activities
plt.figure(figsize=(8, 5))
plt.pie(activities_df["Count"], labels=activities_df["Activity"], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues", len(activities_df)))
plt.title("Internet Activities of Respondents")
plt.show()

# Cyber threats encountered
cyber_threats = {
    "Spear Phishing": 17,
    "Weak Login Credentials": 42,
    "Adware": 32,
    "Social Engineering": 22,
    "Email Spoofing": 65,
    "Other": 21
}

# Convert to DataFrame
threats_df = pd.DataFrame(list(cyber_threats.items()), columns=["Threat Type", "Count"])

# Create a bar plot for cyber threats
plt.figure(figsize=(8, 5))
sns.barplot(x="Count", y="Threat Type", data=threats_df, palette="Reds_r")
plt.xlabel("Number of Respondents")
plt.ylabel("Type of Cyber Threat")
plt.title("Cyber Threats Encountered by Respondents")
plt.show()
```

```python
# Creating a bar chart for Cybersecurity Awareness Levels by Age Group

# Define data for cybersecurity awareness levels by age group
age_groups = ["18-24", "25-34", "35-44", "45+"]
somewhat_aware = [40, 55, 70, 75]  # Percentage of respondents somewhat aware
highly_aware = [10, 20, 30, 35]  # Percentage of respondents highly aware
not_aware = [50, 25, 10, 5]  # Percentage of respondents not very aware or unaware

# Convert to DataFrame for visualization
awareness_data = pd.DataFrame({
    "Age Group": age_groups,
    "Somewhat Aware": somewhat_aware,
    "Highly Aware": highly_aware,
    "Not Aware": not_aware
})

# Plot the bar chart
plt.figure(figsize=(10, 6))
awareness_data.set_index("Age Group").plot(kind="bar", stacked=True, colormap="Blues", figsize=(10, 6))
plt.xlabel("Age Group")
plt.ylabel("Percentage of Respondents")
plt.title("Cybersecurity Awareness Levels by Age Group")
plt.legend(title="Awareness Level")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the figure
awareness_chart_path = "/mnt/data/cybersecurity_awareness_by_age.png"
plt.savefig(awareness_chart_path, dpi=300)
plt.show()

# Provide the download link
awareness_chart_path
```

```python
# Creating a clustered bar chart for Cyber Threat Exposure by Occupation

# Define data for cyber threats encountered by different occupations
occupations = ["Students", "IT Professionals", "Business Employees", "General Users"]
phishing = [60, 30, 40, 50]  # Percentage of each group reporting phishing exposure
malware = [45, 35, 30, 40]  # Percentage reporting malware exposure
credential_theft = [20, 50, 35, 25]  # Percentage reporting credential theft exposure

# Convert to DataFrame for visualization
cyber_threats_df = pd.DataFrame({
    "Occupation": occupations,
    "Phishing": phishing,
    "Malware": malware,
    "Credential Theft": credential_theft
})

# Set occupation as index
cyber_threats_df.set_index("Occupation", inplace=True)

# Plot the clustered bar chart
plt.figure(figsize=(10, 6))
cyber_threats_df.plot(kind="bar", colormap="Reds", figsize=(10, 6))
plt.xlabel("Occupation")
plt.ylabel("Percentage of Respondents")
plt.title("Cyber Threat Exposure by Occupation")
plt.legend(title="Threat Type")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the figure
cyber_threat_chart_path = "/mnt/data/cyber_threat_exposure_by_occupation.png"
plt.savefig(cyber_threat_chart_path, dpi=300)
plt.show()

# Provide the download link
cyber_threat_chart_path
```
